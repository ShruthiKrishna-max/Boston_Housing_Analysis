```python
import pandas as pd
data = pd.read_csv("BostonHousing.csv")
data.head()
```

```python
print("Shape of dataset:", data.shape)

data.info()

data.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(data['medv'], bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of House Prices (MEDV)")
plt.xlabel("Median Value of Houses")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
sns.boxplot(y=data['medv'])
plt.title("Boxplot of Median House Prices (MEDV)")
plt.ylabel("Median Value of Houses")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['rm'], data['medv'], alpha=0.7)
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("Median House Value (MEDV)")
plt.title("Relationship Between Rooms and House Prices")
plt.show()
```

#  Analysis of Boston Housing Dataset

The dataset was explored using descriptive statistics and visualizations to understand the distribution and patterns in housing prices.


The histogram of the median house value (MEDV) shows that most house prices are concentrated in the mid-range, indicating that the majority of houses fall within a moderate price category. A few higher values are observed, suggesting the presence of expensive properties, which slightly skew the distribution.


The boxplot of house prices highlights the spread of the data and reveals a few outliers at the higher end. These outliers may represent premium residential areas or luxury housing segments. The median value lies closer to the lower half of the range, indicating a slight right-skew in the data.


The scatter plot between the average number of rooms (RM) and median house price (MEDV) shows a clear positive relationship. As the number of rooms increases, house prices tend to increase as well. This suggests that the size of the house is an important factor influencing its value.


Overall, the dataset shows meaningful patterns and relationships. The visualizations help identify trends, variability, and outliers, making them useful for further analysis or predictive modeling.
